```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Create a synthetic time-series -----------------------------
np.random.seed(0)
n_periods = 120    # e.g., 10 years of monthly data
t = np.arange(n_periods)

# A simple seasonal pattern + slight upward trend + white noise
seasonal = 10 * np.sin(2 * np.pi * t / 12)
trend = 0.1 * t
noise = np.random.normal(scale=2, size=n_periods)

series = 50 + trend + seasonal + noise

# --- 2. Plot the series --------------------------------------------
plt.figure(figsize=(10, 4))
plt.plot(t, series)
plt.title("Time Series")
plt.xlabel("Time (period index)")
plt.ylabel("Value")
plt.tight_layout()
plt.show()
```
