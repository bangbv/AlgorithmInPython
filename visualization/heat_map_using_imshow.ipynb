```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.rand(10, 10)  # Generate random data for the heatmap
print(f"Data: {data}")  # Print the shape of the data
plt.imshow(data, cmap='hot', interpolation='nearest')  # Create a heatmap using imshow
plt.colorbar()  # Add a color bar to indicate the scale
plt.show()
```
